function takes inputs a number and prints multiplication table of that number

```python
def multiply(multiplicand,multiplier):
    '''
    this function takes 2 inputs multiplicand and multipler and 
    prints the multiples of multiplicand till multipler in the form of 
    '''
    if(multiplier >0):
        i=0
        while(i<=multiplier):
            print(f'{multiplicand}x{i}={multiplicand*i}')
            i+=1
    elif(multiplier ==0):
        print("the multiplier is almost to zero")
    else:
        i=0
        while(i>=multiplier):
            print(f'{multiplicand}x{i}={(multiplicand*i):.4f}')
            i-=1
        
```

```python
multiplicand=float(input("Enter the multiplicand:"))
multiplier=int(float(input("Enter the multiplier:")))
multiply(multiplicand,multiplier)
```

printing twin primes under 1000

```python
def prime(num):
    '''
    This function takes input number 
    returns number if it is prime  else 0
    '''
    for i in range(2,num):
        if num%i==0:
            return 0
    else:
        return num
```

```python
def twinprime(primelst):
    '''
    This function takes list of numbers 
    returns twin primes in form of tuples
    '''
    twinprime=[]
    for i in primelst:
        for j in primelst:
            if j-i==2:
                lst=(i,j)
                twinprime.append(lst)
            else:
                pass
    return twinprime
            
```

```python
numlist=list(range(1000))
primelst=list(map(prime,numlist))
primelst=[x for x in primelst if x>1]
print((twinprime(primelst)))
```
